```python
#整理基準值
standard
```

```python
standard = standard[['Year' , 'Height' , 'Worth', 'Age' , 'International Matches' , 'Goals']]
standard
```

```python
#使用機器學習把NaN補上數值
from sklearn.impute import SimpleImputer
```

```python
#使用平均值替代NaN
imr = SimpleImputer(missing_values=np.nan, strategy ='mean')
new_standard = imr.fit_transform(standard)
new_standard = pd.DataFrame(new_standard)
new_standard.columns= ['Year' , 'Height' , 'Worth', 'Age' , 'International Matches' , 'Goals']
new_standard
```

```python
#算平均
height = standard['Height'].mean()
worth = standard['Worth'].mean()
age = standard['Age'].mean()
InternationalMatches = standard['International Matches'].mean()
goals = standard['Goals'].mean()
df = pd.DataFrame(data = [[height, worth, age, InternationalMatches , goals]],
    columns = ['Height' , 'Worth', 'Age' , 'International Matches' , 'Goals'])
#最終基準
df
```
